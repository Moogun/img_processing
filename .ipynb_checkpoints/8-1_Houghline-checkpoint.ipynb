```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
img = cv2.imread('t1.png',0)
img = cv2.medianBlur(img,5)


ret,th1 = cv2.threshold(img,127,255,cv2.THRESH_BINARY)
ret,th11 = cv2.threshold(img,150,255,cv2.THRESH_BINARY)
ret,th12 = cv2.threshold(img,180,255,cv2.THRESH_BINARY)
ret,th0 = cv2.threshold(img,90,255,cv2.THRESH_BINARY)

th2 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY,11,2)
th3 = cv2.adaptiveThreshold(img,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY,11,2)
titles = ['Original Image', 'Global Thresholding (v = 127)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img, th1, th2, th3]

plt.figure(figsize=(20,10))
for i in range(4):
    plt.subplot(2,2,i+1)
    plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
```

# Dilation

```python
kernel1 = np.ones((1,1),np.uint8)
kernel3 = np.ones((3,3),np.uint8)
kernel5 = np.ones((5,5),np.uint8)
kernel10 = np.ones((10,10),np.uint8)
kernel20 = np.ones((20,20),np.uint8)

krn = [kernel1, kernel3, kernel5, kernel10, kernel20]
```

```python
fig, axs = plt.subplots(1, 5, figsize=(50,10))
fig.suptitle('kernel')

for i in range(5):
    dilation = cv2.dilate(th3,krn[i],iterations = 1)
    axs[i].imshow(dilation,'gray')
    axs[i].set_title('Axis [ krn' + str(i) + ']')
```

```python
edges2 = cv2.Canny(rng,50,150,apertureSize=5)
plt.imshow(edges2, cmap='gray')
```

```python
kernel3 = np.ones((3,3),np.uint8)
dilation = cv2.dilate(edges2,kernel3,iterations = 1)
plt.imshow(dilation)
```

```python
lines = cv2.HoughLinesP(edges2,1,np.pi/180,100, minLineLength = 1000, maxLineGap=200)
red = (255,0,0)
some = (0,0,255)
for line in lines:
    x1, y1, x2, y2 = line[0]
    if y1 - 50 < y2 < y1 + 50: 
        cv2.line(img, (x1, y1), (x2,y2), red)
    else: 
        cv2.line(img, (x1, y1), (x2,y2), some)

plt.imshow(img)
```

```python
lines = cv2.HoughLinesP(dilation,1,np.pi/180,100, maxLineGap=200)
red = (255,0,0)

for line in lines:
    x1, y1, x2, y2 = line[0]
    cv2.line(img, (x1, y1), (x2,y2), red)

plt.imshow(img)
```

```python
def courtLine(orgImg, img, minLineLength, maxLineGap):
    newImg = orgImg.copy()
    lines = cv2.HoughLinesP(img,1,np.pi/180,100, minLineLength=minLineLength, maxLineGap=maxLineGap)
    red = (255,0,0)

    for line in lines:
        x1, y1, x2, y2 = line[0]
        cv2.line(newImg, (x1, y1), (x2,y2), red)

    plt.imshow(newImg)
```

```python
courtLine(img, dilation, 1000, 200)
```
