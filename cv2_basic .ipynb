```python
import cv2
import numpy as np
```

```python
# 1. placeholder img 
img = np.zeros((3, 3), dtype=np.uint8)
img
```

```python
# 2. Convert from 1 ch to 3 cha
img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)
img
```
