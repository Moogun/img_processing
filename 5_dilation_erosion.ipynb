```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
thr180 = cv2.imread('180.jpg')
thr180.shape
```

```python
img = cv2.cvtColor(thr180, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img, cmap='gray')
```

```python
kernel1 = np.ones((1,1),np.uint8)
kernel3 = np.ones((3,3),np.uint8)
kernel5 = np.ones((5,5),np.uint8)
kernel10 = np.ones((10,10),np.uint8)
kernel20 = np.ones((20,20),np.uint8)

krn = [kernel1, kernel3, kernel5, kernel10, kernel20]
```

# Kernel

```python
fig, axs = plt.subplots(1, 5, figsize=(50,10))
fig.suptitle('kernel')

for i in range(5):
    dilation = cv2.dilate(thr180,krn[i],iterations = 1)
    axs[i].imshow(dilation)
    axs[i].set_title('Axis [ krn' + str(i) + ']')
```

```python
iterations = [ 1, 2, 3, 5, 10]
```

# Kernel + Opening

```python
opening = cv2.morphologyEx(thr180, cv2.MORPH_OPEN, kernel3)
plt.imshow(opening, cmap='gray')
```

```python
fig, axs = plt.subplots(1, 5, figsize=(50,10))
fig.suptitle('kernel + opening')

for i in range(5):
    dilation = cv2.dilate(thr180, kernel3,iterations = i)
    opening = cv2.morphologyEx(dilation, cv2.MORPH_OPEN, kernel3)
    axs[i].imshow(dilation)
    axs[i].set_title('Axis [ iteration' + str(i) + ']')
```

# Closing 

```python
closing = cv2.morphologyEx(img, cv2.MORPH_CLOSE, kernel3)
plt.imshow(closing, cmap='gray')
```

# Iterations

```python
fig, axs = plt.subplots(1, 5, figsize=(50,10))
fig.suptitle('iterations')

for i in range(5):
    dilation = cv2.dilate(thr180, kernel3,iterations = i)
    axs[i].imshow(dilation)
    axs[i].set_title('Axis [ iteration' + str(i) + ']')
```

# Erosion

```python
fig, axs = plt.subplots(1, 5, figsize=(50,10))
fig.suptitle('erosion')

for i in range(5):
    erosion = cv2.erode(thr180,krn[i],iterations = 1)
    axs[i].imshow(erosion)
    axs[i].set_title('Axis [ krn' + str(i) + ']')
```
